```python
import pandas as pd

# Load each CSV file
movies_metadata = pd.read_csv('/content/movies_metadata.csv')
ratings = pd.read_csv('/content/ratings.csv')
credits = pd.read_csv('/content/credits.csv')
keywords = pd.read_csv('/content/keywords.csv')

# Preview the first few rows of each DataFrame
print("Movies Metadata:")
display(movies_metadata.head())

print("\nRatings:")
display(ratings.head())

print("\nCredits:")
display(credits.head())

print("\nKeywords:")
display(keywords.head())
```

```python
# Drop irrelevant columns
movies_metadata = movies_metadata.drop(['adult', 'homepage', 'tagline', 'video'], axis=1)

# Drop rows with crucial null values
movies_metadata = movies_metadata.dropna(subset=['title', 'genres', 'release_date'])

# Display updated info
movies_metadata.info()
```

```python
import numpy as np

# Convert budget, revenue, and popularity to numeric
movies_metadata['budget'] = pd.to_numeric(movies_metadata['budget'], errors='coerce')
movies_metadata['revenue'] = pd.to_numeric(movies_metadata['revenue'], errors='coerce')
movies_metadata['popularity'] = pd.to_numeric(movies_metadata['popularity'], errors='coerce')

# Convert timestamp to datetime
ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s')

# Display the first few rows to verify changes
print("Updated Movies Metadata:")
display(movies_metadata.head())

print("\nUpdated Ratings:")
display(ratings.head())
```

```python
import ast

# Extract genres as a list of names
def extract_genres(genre_str):
    genres = []
    for g in ast.literal_eval(genre_str):
        genres.append(g['name'])
    return genres

movies_metadata['genres'] = movies_metadata['genres'].apply(extract_genres)
display(movies_metadata[['title', 'genres']].head())
```

```python
# Extract the top 3 actors
def extract_cast(cast_str):
    cast = []
    for i, actor in enumerate(ast.literal_eval(cast_str)):
        if i < 3:  # Top 3 actors
            cast.append(actor['name'])
    return cast

# Extract director's name
def extract_director(crew_str):
    for member in ast.literal_eval(crew_str):
        if member['job'] == 'Director':
            return member['name']
    return np.nan

credits['cast'] = credits['cast'].apply(extract_cast)
credits['director'] = credits['crew'].apply(extract_director)

display(credits[['cast', 'director']].head())
```

```python
# Extract keywords as a list
def extract_keywords(keyword_str):
    keywords = []
    for kw in ast.literal_eval(keyword_str):
        keywords.append(kw['name'])
    return keywords

keywords['keywords'] = keywords['keywords'].apply(extract_keywords)
display(keywords.head())
```

```python
# Convert 'id' column in both DataFrames to the same data type (e.g., int64)
movies_metadata['id'] = pd.to_numeric(movies_metadata['id'], errors='coerce').astype('Int64')
credits['id'] = pd.to_numeric(credits['id'], errors='coerce').astype('Int64')

# Now perform the merge
movies_df = movies_metadata.merge(credits[['id', 'cast', 'director']], on='id')
movies_df = movies_df.merge(keywords, on='id')

# Display merged DataFrame
display(movies_df.head())
```

```python
# Select relevant features
movies_df = movies_df[['title', 'genres', 'cast', 'director', 'keywords', 'overview', 'release_date', 'vote_average', 'vote_count']]

# Display the final cleaned DataFrame
display(movies_df.head())
```

```python
# Import the necessary class
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert lists to strings for TF-IDF, handling NaN values
movies_df['genres'] = movies_df['genres'].apply(lambda x: ' '.join(x) if isinstance(x, list) else '')
movies_df['cast'] = movies_df['cast'].apply(lambda x: ' '.join(x) if isinstance(x, list) else '')
movies_df['keywords'] = movies_df['keywords'].apply(lambda x: ' '.join(x) if isinstance(x, list) else '')
movies_df['director'] = movies_df['director'].fillna('').astype(str) # Handle NaN in 'director'

# Combine relevant features into a single 'combined_features' column
movies_df['combined_features'] = movies_df['genres'] + ' ' + movies_df['keywords'] + ' ' + movies_df['cast'] + ' ' + movies_df['director']

# Fill NaN values in 'combined_features' with an empty string
movies_df['combined_features'] = movies_df['combined_features'].fillna('')

# Initialize TF-IDF Vectorizer
tfidf = TfidfVectorizer(stop_words='english')

# Vectorize the combined features
tfidf_matrix = tfidf.fit_transform(movies_df['combined_features'])

print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute Cosine Similarity Matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Display similarity matrix shape
print("Cosine Similarity Matrix Shape:", cosine_sim.shape)
```

```python
# Create a Series mapping movie titles to DataFrame indices
movie_indices = pd.Series(movies_df.index, index=movies_df['title']).drop_duplicates()

# Function to get movie recommendations
def get_recommendations(title, cosine_sim=cosine_sim, num_recommendations=10):
    # Get the index of the movie that matches the title
    idx = movie_indices[title]

    # Get the pairwise similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the most similar movies
    sim_scores = sim_scores[1:num_recommendations+1]  # Exclude itself

    # Get the movie indices
    movie_indices_list = [i[0] for i in sim_scores]

    # Return the top N most similar movies
    return movies_df['title'].iloc[movie_indices_list]

# Test the recommendation system
print("Movies similar to 'Toy Story':")
print(get_recommendations('Toy Story'))
```

```python
from surprise import Dataset, Reader
import pandas as pd

# Load the ratings dataset again
ratings_path = '/content/ratings.csv'
ratings = pd.read_csv(ratings_path)

# Load data into Surprise's Dataset format
reader = Reader(rating_scale=(0.5, 5.0))
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)

# Split data into training and testing sets
from surprise.model_selection import train_test_split
trainset, testset = train_test_split(data, test_size=0.25)

print("Data re-prepared for Collaborative Filtering using KNN!")
```

```python
from surprise import KNNBasic

# Define the similarity measure
sim_options = {
    'name': 'cosine',      # Use cosine similarity
    'user_based': False    # Set to False for Item-Item, True for User-User
}

# Initialize the KNN model
knn_model = KNNBasic(sim_options=sim_options)

# Train the model on the training set
knn_model.fit(trainset)

print("KNN Model trained successfully!")
```

```python
# Get a list of all movie IDs
all_movie_ids = ratings['movieId'].unique()

# Function to get top N movie recommendations for a user
def get_knn_recommendations(user_id, num_recommendations=10):
    # Get a list of movies already rated by the user
    rated_movies = ratings[ratings['userId'] == user_id]['movieId'].tolist()

    # Predict ratings for all movies not yet rated by the user
    movie_scores = []
    for movie_id in all_movie_ids:
        if movie_id not in rated_movies:
            # Predict rating
            pred = knn_model.predict(user_id, movie_id)
            movie_scores.append((movie_id, pred.est))

    # Sort by predicted rating and select top N
    movie_scores.sort(key=lambda x: x[1], reverse=True)
    top_movies = movie_scores[:num_recommendations]

    # Return the top N recommended movie IDs
    top_movie_ids = [movie[0] for movie in top_movies]
    return top_movie_ids

# Test the recommendation system
print("\nTop 10 Movie Recommendations for User 1:")
print(get_knn_recommendations(user_id=1))
```
